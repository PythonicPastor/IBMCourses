<h3>The code below is used to scrape job listings from the indeed website:</h3>

```python
import requests
from bs4 import BeautifulSoup as bs
import re
#import os
import webbrowser

def linkPG(x, query):
    url ="https://www.indeed.com/jobs?q="+query+"&sc=0kf%3Aattr(DSQF7)%3B&sort=date&fromage=3&start="+str(x*10)+"&vjk=86d6b8e2ca71d8dd"

    result = requests.get(url)
    
    src = result.content
    soup = bs(src, 'html.parser')
    return soup.find_all('a', id=re.compile('^job_\d+'))

query = input('Search:')
query = query.replace(" ", "%20")
jobs = []
for x in range(10):
    links = linkPG(x, query)
    for i in range(len(links)):
        jobs.append([links[i].text, "https://www.indeed.com" + links[i]['href']])
        print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
        print(jobs[i][0], '\n', jobs[i][1])
        print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~', '\n')
```
