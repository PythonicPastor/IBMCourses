```python
import folium
latLong = [37.72860956589776, -89.21575359246572]
CarbondaleMap= folium.Map(
                        location=latLong,
                        zoom_start=20
                        )
bombHere = folium.map.FeatureGroup()

bombHere.add_child(
                    folium.features.CircleMarker(
                        latLong,
                        radius = 50,
                        color ='red',fill_color='red'
                                                )
                    )
CarbondaleMap.add_child(bombHere)
folium.Marker(latLong,
              popup='Bomb Here'
             ).add_to(CarbondaleMap)
CarbondaleMap
```
